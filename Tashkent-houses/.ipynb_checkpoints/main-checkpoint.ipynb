```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_data(file_path):
    return pd.read_excel(file_path)
```

```python
def main(dataset):
    df = dataset
    return df
dataset = load_data('dataset.xlsx')
main(dataset)
```

```python
def visualize(dataset):
    sns.color_palette("rocket", as_cmap=True)
    sns.relplot(data=dataset, x='lat', y='lng', hue='district', size='size', sizes=(40, 400))
    plt.show()
visualize(dataset)
```
